Crear y guardar una gráfica de pay del año más reciente existente en la base de datos.

```python
# Librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Se cargan el dataset
medallas = pd.read_csv('athlete_events.csv')
medallas.head()
```

```python
# Se selecciona la info de el año mas reciente
medallas_last_year = medallas[medallas['Year'] == medallas['Year'].max()]
medallas_last_year
```

```python
# Se eliminan los nulos de la columna Medal
medallas_last_year = medallas_last_year.dropna()
medallas_last_year
```

```python
# Se agrupan or pais para grafica de Pie
medallas_last_year_pais = medallas_last_year.groupby('NOC').count()['Medal']
medallas_last_year_pais
```

```python
# Se obtienen el Top 10 paises para graficar
medallas_last_year_pais_top10 = medallas_last_year_pais.sort_values(ascending=False).head(10)
medallas_last_year_pais_top10
```

```python
# Se graba la grafica como .png
ax = medallas_last_year_pais_top10.plot.pie(figsize=(10,10), autopct = '% 0.1f %%')
plt.title('Top 10 de Paises Ganadores en el año mas reciente (2016)')
fig = ax.get_figure()
fig.savefig("Grafica Top 10 del 2016.jpg")
```
